```python
import numpy as np
import plotly.graph_objects as go

# Define spherical coordinates
u = np.linspace(0, 2 * np.pi, 100)  # azimuthal angle
v = np.linspace(0, np.pi, 50)       # polar angle

# Parametric equations for a sphere
x = np.outer(np.cos(u), np.sin(v))
y = np.outer(np.sin(u), np.sin(v))
z = np.outer(np.ones(np.size(u)), np.cos(v))

# Create the 3D sphere plot using Plotly
fig = go.Figure(data=[go.Surface(z=z, x=x, y=y, colorscale='Blues')])


# Prepare arrays x, y, z
theta_1,fi_1 = (-0.4, 0.4)
theta_2,fi_2 = (1.3, -0.4)


# if(abs(theta_2 - theta_1) > abs(theta_2 - 2*np.pi - theta_1)):
#     print("temp")
r = 1
t = np.linspace(0, 1, 100)
x_t = (r * np.cos(theta_1) * np.cos(fi_1))*t + (r * np.cos(theta_2) * np.cos(fi_2))*(1-t)
y_t = (r * np.cos(theta_1) * np.sin(fi_1))*t + (r * np.cos(theta_2) * np.sin(fi_2))*(1-t)
z_t = (r * np.sin(theta_1))*t + (r * np.sin(theta_2))*(1-t)

# Parametric line: x(t) = cos(t), y(t) = sin(t), z(t) = t
t = np.linspace(0, 2 * np.pi, 100)

width = np.sqrt(x_t*x_t + y_t*y_t + z_t*z_t)

x_line = x_t*r /(width)
y_line = y_t*r /(width)
z_line = z_t*r /(width)
# Add the parametric line to the plot
fig.add_trace(go.Scatter3d(x=x_line, y=y_line, z=z_line, mode='lines', 
                           line=dict(color='red', width=4), name='Parametric Line'
                           ))

# Update layout for better presentation
fig.update_layout(title="3D Sphere with Parametric Line",
                  scene=dict(
                      xaxis_title='X',
                      yaxis_title='Y',
                      zaxis_title='Z'),
                  autosize=True)

# Show the plot
fig.show()
```
